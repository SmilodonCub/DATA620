### Analyzing Tweets and Tweet Entities: What are people talking about right now?

Creating a basic frequency distribution from the words in tweets

```python
q = '#FreeBritney -is:retweet'
count = 100

search_results = client.search_recent_tweets( query = q,  
                                             tweet_fields=['context_annotations', 'created_at', 'geo', 'entities'], 
                                             place_fields=['place_type', 'geo'], 
                                             expansions='geo.place_id',
                                             max_results=count )
for tweet in search_results.data:
    #print( tweet.text )
    if len( tweet.entities )>0:
        print( tweet.entities['hashtags'][0]['tag'])
```

```python
hashtags = [tweet.entities['hashtags'][0]['tag'] for tweet in search_results.data]
```

```python
from collections import Counter

c = Counter( hashtags )
print( c.most_common()[:10 ])
print()
```

```python
from prettytable import PrettyTable

pt = PrettyTable( field_names = [ 'Hashtag', 'Count' ] )
[ pt.add_row( kv ) for kv in c.most_common()[:10] ]
pt.align[ 'Hashtag' ], pt.align[ 'Count' ] = 'l', 'r'
print( pt )
```

### Lexical Diversity

**lexical diversity** - an expression of the number of unique tokens in the text divided by the number of tokens in the text. can be interpreted as a measure of the depth of detail for a discussion. a measure of understanding of a subject. EX: **comparing effectiveness of social media marketing campaigns** on Twitter. Does lexical Diversity change over time given the launch of a campaign?  

<br>

```python
status_texts = [tweet.text for tweet in search_results.data]
words = [ word for text in status_texts for word in text.split() ]
len(words)
```

```python
def lexical_diversity( tokens ):
    return len( set(tokens) )/len(tokens)

def average_words( statuses ):
    total_words = sum( [len(s.split()) for s in statuses ] )
    return total_words/len(statuses)
```

```python
print( lexical_diversity( hashtags ) )
print( lexical_diversity( words ) )
print( average_words( status_texts ) )
```

```python

q = 'huggingface -is:retweet'
count = 100

search_results = client.search_recent_tweets( query = q,  
                                             tweet_fields=['context_annotations', 'created_at', 'public_metrics', 'entities', 'author_id'], 
                                             place_fields=['place_type', 'geo'], 
                                             max_results=count )
names_list = []
for tweet in search_results.data:
    author_id = tweet.author_id
    author_name = client.get_user( id = author_id, user_fields=['username'])
    names_list.append( author_name[0]['username'] )
```

```python
retweets = [(tweet.public_metrics['retweet_count'], 
             tweet.public_metrics['reply_count'],
            tweet.public_metrics['like_count'],
            tweet.public_metrics['quote_count'],
            tweet.author_id,
            names_list[idx],
            tweet.text) for idx,tweet in enumerate(search_results.data)]
retweets[0:10]
```

```python
field_names = [ 'retweet', 'reply', 'like', 'quote', 'author_id', 'handle', 'tweet_text' ]
pt = PrettyTable( field_names = field_names )
[ pt.add_row(row) for row in sorted( retweets, reverse=True)[:20]]
pt.max_width['tweet_text'] = 25
pt.align='l'
print(pt)
```

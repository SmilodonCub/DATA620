<h2><center><font color='blue'>the Erdős Collaboration Network</font></h2></center>

Paul Erdős (1913-1996) was a Hungarian mathemetician who was particularly proliphic during his time. He has published approximately 1,525 papers with many different authors. The Erdős Collaboration Network is a representation of his body of work where nodes represent authors (including Paul Erdős) and edges represent connections as co-authors on papers.  

The following code will visualize the Erdős Collaboration Network:

```python
# import the networkx library
import networkx as nx
import matplotlib.pyplot as plt
```

Data was downloaded from the [Pajek datasets site](http://vlado.fmf.uni-lj.si/pub/networks/data/) as a .net file

```python
# import the Erdos Collaboration .net file
file = 'Erdos02.net'
G = nx.read_pajek(file)
print(len((G.nodes())))
```

Explore a few of the methods available to [draw networks](https://networkx.org/documentation/stable/reference/drawing.html)

```python
# visualize the network
fig, ax = plt.subplots(figsize=(15, 8))
nx.draw_circular(G, node_size=11)
plt.show()
```

```python
# visualize the network
fig, ax = plt.subplots(figsize=(15, 8))
nx.draw_random(G, node_size=11)
plt.show()
```

```python
# visualize the network
fig, ax = plt.subplots(figsize=(15, 8))
nx.draw_spectral(G, node_size=11)
plt.show()
```

```python
# visualize the network
fig, ax = plt.subplots(figsize=(15, 8))
nx.draw_spring(G, node_size=11)
plt.show()
```
